```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fileimport = "database_gas.csv"
```

```python
data=pd.read_csv(fileimport)
data.head(5)
```

```python
data=data.loc[:, ['temperature','humidity']]
plt.scatter(data.temperature,data.humidity)
```

```python
data=np.array(data)
```

```python
K=5
m=3

U=np.random.rand(data.shape[0],K)
U/=np.sum(U, axis=1)[:,np.newaxis]
U
```

```python
def calculate_centroids(data, K, U, m):
    Centroids = np.zeros((K,data.shape[1]))
    for i in range(K):
        Centroids[i,:] = np.sum((U[:,i]**m)[:,np.newaxis]*data, axis=0)/np.sum(U[:,i]**m)
    return Centroids

Centroids = calculate_centroids(data, K, U, m)
Centroids    
```

```python
U_new = np.zeros((data.shape[0],K))
for i in range(K):
    U_new[:,i] = np.linalg.norm(data-Centroids[i,:], axis=1)

U_new=1/ (U_new**(2/(m-1)) * np.sum((1/U_new) ** (2/(m-1)), axis=1)[: ,np.newaxis])
U_new
```

```python
def calculate_membership (data, Centroids, K , m):
    U_new=np.zeros((data.shape[0],K))
    for i in range (K):
        U_new[:,i]=np.linalg.norm(data-Centroids[i,:],axis=1)

    U_new=1/ (U_new ** (2/(m-1)) * np.sum((1/U_new) ** (2/(m-1)) , axis=1 )[:,np.newaxis] )
    return U_new

calculate_membership(data, Centroids, K , m)
```

```python
labels=np.argmax(U_new,axis=1)
labels
```

```python
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, palette='viridis')
```

```python
iteracoes=90

for i in range (iteracoes):
    Centroids = calculate_centroids(data, 5, U, 3)
    U_new = calculate_membership(data, Centroids, 5, 3)

    labels=np.argmax(U_new,axis=1)

    if np.linalg.norm(U_new-U) < 0.0010:
        break
    U = U_new

    labels=np.argmax(U_new,axis=1)
```

```python
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, palette='viridis')
```
